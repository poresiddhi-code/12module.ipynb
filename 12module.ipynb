What is a Module

Modules refer to a file containing Python statements and definitions.

Modules enable you to split parts of your program in different files for easier maintenance and better performance.

There are thousands of modules in Python. Some are built-in, others are user-created.

We'll explore one module here called math.


Math Module

Modules need to be imported.

import math

When you run the above line, all the functions that are written in math module are accessible for us to use.

```python
# As usual, let's see what this module is about with help()

help(math)
```

```python
# so we got list of all the functions that are written/stored in this module.
# let's try using sqrt

math.sqrt(4)           # if we simply right sqrt(), we'll get error. why. Because sqrt() is not default function (like say len()). sqrt() function belongs to math module
```

```python
math.sqrt(5)
```

```python
math.sqrt(6)
```

```python
math.sqrt(2)
```

```python
math.sqrt(7)
```

```python
# But if you import sqrt() specifically, then you can access it just like that.
# e.g.

from math import sqrt       # use ctrl + tab to get suggestions after import
sqrt(4)                     # we can use sqrt() now because in the above line, we are explicitly importing sqrt() function
```

Date module

Let's explore datetime module to get current date.

Since each module comes with its own function, you need to explore those functions as needed.

One way is to use help()

Other is google.



```python
from datetime import date
```

```python
print(date.today())
```

```python
today = date.today()
print(today)
```

```python
print(today.day, today.month, today.year)       # ctrl+tab after today.
```

```python
from datetime import datetime
```

```python
print(datetime.now())         # This gives GMT time (IST = GMT + 5.30)
```

```python
print(datetime.now().strftime("%Y-%m-%d %H:%M:%S"))
```
